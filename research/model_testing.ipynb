```python
from tensorflow.keras.models import load_model
```

```python
model = load_model('data/model_checkpoint/Chicken_fecal_with_callbacks.keras')
```

# test the model

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('data/data_ingestion/extracted_files/Chicken-fecal-images/Coccidiosis/cocci.10.jpg')
```

```python
plt.imshow(img)
plt.show()
```

```python
img = img/255.0
img = img.reshape(1,224,224,3)
```

```python
test_pred = model.predict(img)
```

```python
test_pred
```

```python
aa = np.argmax(test_pred)

if aa == 0:
    print("Coccidiosis")
else:
    print("Healthy")
```
